```python
import requests
import bs4 as BeautifulSoup

from selenium import webdriver
from selenium.webdriver import ActionChains

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

URL = 'https://www.sebi.gov.in/sebiweb/other/OtherAction.do?doMfd=yes&type=4'

driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get(URL)
```

```python
import pandas as pd
from datetime import datetime

from selenium.webdriver.common.by import By

list_of_tables = driver.find_elements(By.TAG_NAME,"tbody")

list_of_rows = list_of_tables[0].find_elements(By.TAG_NAME,"tr")

list_of_dictionaries = []

n_rows = len(list_of_rows)

i = 0
j = 0

while i in range(0,n_rows):
    
    temp_dictionary = {}
    selected_row = list_of_rows[i]
    
    cell_data = selected_row.find_elements(By.TAG_NAME,"td")
    
    new_date = cell_data[0].text
    new_date = datetime.strptime(new_date, '%d %b, %Y')
    new_date = datetime.date(new_date)
    
    temp_dictionary['DATE'] = new_date
    temp_dictionary['DEBT/EQUITY'] = cell_data[1].text
    temp_dictionary['GROSS PURCHASES (RS.CRORE)'] = cell_data[2].text
    temp_dictionary['GROSS SALES (RS. CRORE)'] = cell_data[3].text
    temp_dictionary['NET INVESTMENT (RS. CRORE)'] = cell_data[4].text
    
    list_of_dictionaries.append(temp_dictionary)
    temp_dictionary = {}
    
    i+=1
    
    selected_row = list_of_rows[i]
    
    cell_data = selected_row.find_elements(By.TAG_NAME,"td")
    
    temp_dictionary['DATE'] = new_date
    temp_dictionary['DEBT/EQUITY'] = cell_data[0].text
    temp_dictionary['GROSS PURCHASES (RS.CRORE)'] = cell_data[1].text
    temp_dictionary['GROSS SALES (RS. CRORE)'] = cell_data[2].text
    temp_dictionary['NET INVESTMENT (RS. CRORE)'] = cell_data[3].text
    
    list_of_dictionaries.append(temp_dictionary)
    
    i+=1
```

```python
df = pd.DataFrame(list_of_dictionaries)
df
```

```python
Equity_df = df[df['DEBT/EQUITY'] == 'Equity']
Equity_df = Equity_df.reset_index(drop = True)
Equity_df
```

```python
Debt_df = df[df['DEBT/EQUITY'] == 'Debt']
Debt_df = Debt_df.reset_index(drop = True)
Debt_df
```
